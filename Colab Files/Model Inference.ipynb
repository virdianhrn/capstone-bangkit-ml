```python
from transformers import BertTokenizer, TFBertModel
import tensorflow as tf
import keras
from keras import layers, metrics
import pandas as pd
import numpy as np
import math
import string
import unicodedata
```

```python
def get_max_len_batch(texts):
  max_len = 0
  for text in texts:
    input_ids = tokenizer(text, add_special_tokens=True)['input_ids']
    max_len = max(max_len, len(input_ids))

  return min(2 ** math.ceil(math.log2(max_len)), 2048)
```

```python
def encode(text, max_len):
    encoded_dict = tokenizer(text, add_special_tokens = True, max_length = max_len,
                             padding='max_length', return_attention_mask = True,
                             truncation=True, return_tensors = 'tf')

    input_ids = encoded_dict['input_ids']
    attention_masks = encoded_dict['attention_mask']
    return (input_ids, attention_masks)
```

```python
def split_input(input_id, attention_mask):
    return {'input_ids': input_id, 'attention_mask': attention_mask}
```

```python
def dataset_preperation(text, max_len):
  features_tensor = encode(text.values.tolist(), max_len)
  dataset = tf.data.Dataset.from_tensor_slices(features_tensor)
  dataset = dataset.map(split_input)

  return (dataset.batch(1))
```

```python
def preprocess(text):
    text = unicodedata.normalize('NFKD', text)
    text = text.replace('\\t'," ").replace('\\n'," ").replace('\\u'," ")
    text = text.replace('\\',"").replace('\n\n'," ").replace("KOMPAS.com -", "")
    return text
```

## Subjectivity

```python
def subj_predict_batch(subjectivity_model, subj_csv_path, subj_col_name):
  df_subjectivity = pd.read_csv(subj_csv_path)
  df_subjectivity[subj_col_name] = df_subjectivity[subj_col_name].astype(str).apply(preprocess)

  features = df_subjectivity[subj_col_name]
  max_len_subjectivity = get_max_len_batch(features)

  dataset_subj = dataset_preperation(features, max_len_subjectivity)
  return subjectivity_model.predict(dataset_subj)
```
